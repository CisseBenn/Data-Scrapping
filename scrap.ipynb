```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
from textblob import TextBlob
```

```python
url = "https://www.kilimall.co.ke/category/bags?id=1&form=category&page=1"
response = requests.get(url)
response
```

```python
sourcecode = BeautifulSoup(response.content, "html.parser")
sourcecode
```

```python
product_name = sourcecode.find_all(class_="product-title")
product_price = sourcecode.find_all(class_="product-price")
product_price
```

```python
productname = []
for item in product_name:
    productname.append(item.text)
productname
```

```python
productprice = []
for x in product_price:
    productprice.append(x.text)
productprice
```

```python
data = {
    "Product Name": productname,
    "Price": productprice
}

df = pd.DataFrame(data)
print(df)
```

```python
testurl = "https://play.google.com/store/apps/details?id=com.adobe.lrmobile&hl=en"

response1 = requests.get(testurl)

response1
```

```python
content = response1.content
test_sourcecode = BeautifulSoup(content, 'html.parser')

test_sourcecode
```

```python
reviews = []
review = test_sourcecode.find_all(class_=["h3YV2d"])

for rev in review:
    reviews.append(rev.text)

reviews
```

```python
# Convert reviews list to a single string
text = " ".join([review for review in reviews])  

sentiment = TextBlob(text).sentiment

print(f"Polarity: {sentiment.polarity},  Subjectivity: {sentiment.subjectivity}")
```

```python
comment_sentiment = []

for comment in reviews:
    sentiment = TextBlob(comment).sentiment
    sentimentpolarity = sentiment.polarity
    comment_sentiment.append(sentimentpolarity)
comment_sentiment
```

```python
newcompletesentiment = []

for c in comment_sentiment:
    if c == 0:
        sent = "Neutral"
        newcompletesentiment.append(sent)
    elif c > 0:
        sent = "Positive"
        newcompletesentiment.append(sent)
    elif c < 0:
        sent = "Negative"
        newcompletesentiment.append(sent)

newcompletesentiment
```

```python
names = ["James", "Matthew", "Jonas", "Mary", "Andrew"]
feedback = ["This item is broken", "The product is awesome", "Delivery time was fast but the package was missing some products", "Perfect weather to go swimming", "Provide feedback on the product"]
```

```python
feedbacks = []

for f in feedback:
    sentiment = TextBlob(f).sentiment
    sentpolarity = sentiment.polarity
    feedbacks.append(sentpolarity)

feedbacks
```

```python
finalfeedback = []

for feed in feedbacks:
    if feed == 0:
        sent = "Neutral feedback"
        finalfeedback.append(sent)

    elif feed > 0:
        sent = "Positive Feedback"
        finalfeedback.append(sent)
    elif feed < 0:
        sent = "Negative feedback"
        finalfeedback.append(sent)

finalfeedback
```

```python
df = pd.DataFrame ({
    "Names" : names,
    "Feedback" : finalfeedback 
})

df
```
