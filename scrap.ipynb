```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = "https://www.kilimall.co.ke/category/bags?id=1&form=category&page=1"
response = requests.get(url)
response
```

```python
sourcecode = BeautifulSoup(response.content, "html.parser")
sourcecode
```

```python
product_name = sourcecode.find_all(class_="product-title")
product_price = sourcecode.find_all(class_="product-price")
product_price
```

```python
productname = []
for item in product_name:
    productname.append(item.text)
productname
```

```python
productprice = []
for x in product_price:
    productprice.append(x.text)
productprice
```

```python
data = {
    "Product Name": productname,
    "Price": productprice
}

df = pd.DataFrame(data)
print(df)
```
