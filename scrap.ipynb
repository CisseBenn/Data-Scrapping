```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = "https://www.kilimall.co.ke/category/bags?id=1&form=category&page=1"
response = requests.get(url)
response
```

```python
sourcecode = BeautifulSoup(response.content, "html.parser")
sourcecode
```

```python
product_name = sourcecode.find_all(class_="product-title")
product_price = sourcecode.find_all(class_="product-price")
product_price
```

```python
productname = []
for item in product_name:
    productname.append(item.text)
productname
```

```python
productprice = []
for x in product_price:
    productprice.append(x.text)
productprice
```

```python
data = {
    "Product Name": productname,
    "Price": productprice
}

df = pd.DataFrame(data)
print(df)
```

```python
url = "https://www.ebay.com/globaldeals/sch?_nkw=iphone+12+pro+max&_sacat=-1&_from=R40&_trksid=p2344380.m570.l1311&_sop="
```

```python
response = requests.get(url)
response
```

```python
sourcecode1 = BeautifulSoup(response.content, "html.parser")
sourcecode1
```

```python
phonenames = sourcecode1.find_all(class_="ebayui-ellipsis-2")
phonenames
```

```python
phoneprices = sourcecode1.find_all(class_="first")
phoneprices
```

```python
phone_names = []
for x in phonenames:
    phone_names.append(x.text)
phone_names
```

```python
phone_prices = []

for y in phoneprices:
    phone_prices.append(y.text)
    
phone_prices
```

```python
phonedata = {
    "Phone Name": phone_names,
    "Phone Price": phone_prices
}

phone_df = pd.DataFrame(phonedata)
phonedata
```
