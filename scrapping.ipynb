```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://jiji.co.ke/cars?filter_attr_1_make=Audi"
```

```python
response = requests.get(url)
response
```

```python
sourcecode = BeautifulSoup(response.content,"html.parser")
print(sourcecode)
```

```python
#Getting data from the source code above
carprices = sourcecode.find_all(class_="qa-advert-price")
carprices
```

```python
carnames = sourcecode.find_all(class_="b-advert-title-inner qa-advert-title b-advert-title-inner--div")
carnames
```

```python
#Extracting plain text from the data above
carprices_array = []
for item in carprices:
    carprices_array.append(item.text)
    
```

```python
carprices_array
```

```python
carnames_array = []
for names in carnames:
    carnames_array.append(names.text)
```

```python
carnames_array
```

```python
zip_list = list(zip(carnames_array, carprices_array))

df1 = pd.DataFrame(zip_list)
df1
```

```python
#Creating a dataframe
data = {
    "Car Names": carnames_array,
    "Prices": carprices_array
}
df = pd.DataFrame(data)
print(df)
```
